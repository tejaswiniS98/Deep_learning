```
#Activation functions
#plot the graph for different activation function
import numpy as np
import matplotlib.pyplot as plt
```

```
#linear function
x = np.arange(-2 , 2, 0.1)
def linear(x):
  a=[]
  for item in x:
    a.append(item)
  return a 
y = linear(x)
print(y)  
```

```
#plot the grah of linear function
plt.plot(x , y,'--','b')
```

```
#define x-axis and y-axis
plt.axhline(0,color = 'r')
plt.axvline(0,color = 'r')
plt.plot(x,y,color = 'b')
plt.show()
```

```
#sigmoid function
#define sigmoid function
def sigmoid(x):
  return(1/1+np.exp(-x))
```

```
#plot the graph
x = np.arange(-2,2,0.1)
y = sigmoid(x)
plt.axhline(0,color = 'black')
plt.axvline(0,color = 'r')
plt.plot(x,y,'o',color='b')
plt.show()
```

```
#tanh function
plt.axhline(0, color = 'r')
plt.axvline(0 ,color = 'r')

plt.plot(x,np.tanh(x))
plt.show()
```

```
plt.axhline(0,color = 'black')
plt.axvline(0,color = 'r')
plt.plot(x,np.tanh(x))
plt.grid()
plt.show()
```

```
#define rectified linear unit
def relu(x):
  a = []
  for item in x:
    if item > 0:
      a.append(item)
    else:
      a.append(0)
  return a     
```

```
#plot the graph of relu activation function
y = relu(x)
print(y)
plt.axhline(0,color ='g')
plt.axvline(0,color = 'g')
plt.plot(x,y)
plt.show()
```

```
#Exponential Linear Unit 
def eLU(x, alpha=1):
  return np.array([xi if xi>0 else alpha*(np.exp(xi)-1) for xi in x ])
plt.axhline(0,color='r') 
plt.axvline(0,color='g') 
plt.plot(x, eLU(x))
```

```
#LogSigmoid activation function
def logsigmoid(x, lam=1):
  return np.log(1/(1+np.exp(-lam*x)))
```

```

plt.axhline(0,color='g') 
plt.axvline(0,color='g') 
plt.plot(x, logsigmoid(x))
```
